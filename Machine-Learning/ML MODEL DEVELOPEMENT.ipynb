### Model Evaluation

Let's compute various evaluation metrices now
- Accuracy Score

- Precision - Recall curve 

```python
from sklearn.metrics import confusion_matrix 
from sklearn.metrics  import classification_report 
from sklearn.metrics  import accuracy_score
from sklearn.metrics import ConfusionMatrixDisplay as plot_confusion_matrix
from sklearn.metrics  import roc_auc_score 
from sklearn.metrics  import RocCurveDisplay as plot_roc_curve 
from sklearn.metrics import precision_recall_curve 
from sklearn.metrics  import PrecisionRecallDisplay as plot_precision_recall_curve
```

```python
from sklearn.metrics import (
    accuracy_score,
    roc_auc_score,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay,
    RocCurveDisplay,
    PrecisionRecallDisplay
)
import matplotlib.pyplot as plt

def compute_evaluation_metric(model, x_test, y_actual, y_predicted, y_predicted_prob):
    print("\nAccuracy Score :", accuracy_score(y_actual, y_predicted))
    print("\nAUC - ROC Score :", roc_auc_score(y_actual, y_predicted_prob))
    print("\nClassification Report:\n", classification_report(y_actual, y_predicted))
    print("\nConfusion Matrix:\n", confusion_matrix(y_actual, y_predicted))
    
    print("\nConfusion Matrix Plot:\n")
    ConfusionMatrixDisplay.from_predictions(y_actual, y_predicted, cmap='plasma', values_format='d')
    plt.show()

    print("\nAUC - ROC Curve:\n")
    RocCurveDisplay.from_estimator(model, x_test, y_actual)
    plt.show()

    print("\nPrecision - Recall Curve:\n")
    PrecisionRecallDisplay.from_estimator(model, x_test, y_actual)
    plt.show()
```

```python
# Compute Evaluation Metric
compute_evaluation_metric(rfc, X_test, y_test, y_pred_rfc, y_prob_pred_rfc)
```

# Improving Fraud Detection Model Performance

The goal in fraud detection is typically to maximize recall (catching all frauds) while maintaining acceptable precision (minimizing false alarms). Let's implement several techniques to improve our model's performance, with special focus on recall.

## Techniques we'll use:
1. Address class imbalance with SMOTE
2. Feature engineering with interaction features
3. Cost-sensitive learning
4. Threshold optimization
5. Ensemble methods

## 1. Handling Class Imbalance with SMOTE

Class imbalance is a common issue in fraud detection problems, as fraudulent transactions typically represent a small minority of all transactions. SMOTE (Synthetic Minority Over-sampling Technique) creates synthetic examples of the minority class to balance the dataset.

```python
# Install imbalanced-learn if not already installed
# Fix dependencies and install correct versions
# !pip install -q scikit-learn==1.4.1 imbalanced-learn==0.10.1
# Full dependency resolution
!pip install -U scikit-learn==1.2.2
!pip install -U lightgbm==3.3.5
!pip install -U numpy==1.24.3
!pip install -U pandas==2.0.3
!pip install -U imbalanced-learn==0.10.1
# !pip install -q imbalanced-learn
# !pip install -q rich==13.7.1
# !pip install -q toolz==0.11.2


from imblearn.over_sampling import SMOTE
from collections import Counter

# Check class distribution before resampling
print("Original class distribution:")
print(Counter(y_train))

# Apply SMOTE to balance the classes
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Check class distribution after resampling
print("Resampled class distribution:")
print(Counter(y_train_resampled))
```

## 2. Feature Engineering - Creating Interaction Features

Let's create interaction features between important variables identified by our models.

```python
# Create interaction features between important variables
X_train_enhanced = X_train_resampled.copy()
X_test_enhanced = X_test.copy()

# Interaction between card1 and card2
X_train_enhanced['card1_card2'] = X_train_resampled['card1'] * X_train_resampled['card2']
X_test_enhanced['card1_card2'] = X_test['card1'] * X_test['card2']

# Interaction between addr1 and addr2
X_train_enhanced['addr1_addr2'] = X_train_resampled['addr1'] * X_train_resampled['addr2']
X_test_enhanced['addr1_addr2'] = X_test['addr1'] * X_test['addr2']

# Ratio features
X_train_enhanced['card1_to_addr1'] = X_train_resampled['card1'] / (X_train_resampled['addr1'] + 1)  # Adding 1 to avoid division by zero
X_test_enhanced['card1_to_addr1'] = X_test['card1'] / (X_test['addr1'] + 1)

# Check the new features
print(X_train_enhanced.columns)
print(X_test_enhanced.shape)
```

## 3. Cost-Sensitive Learning with LightGBM

We'll use the `class_weight` parameter to make the model more sensitive to the minority class (fraud).

```python
%%time
# Define the model with cost-sensitive learning
# Higher weight for the positive class (fraud)
lgbc_cost_sensitive = LGBMClassifier(
    random_state=42, 
    max_depth=20, 
    n_estimators=100, 
    num_leaves=256,
    class_weight={0: 1, 1: 10},  # Giving more weight to fraud class
    n_jobs=-1
)

# Train the model
lgbc_cost_sensitive.fit(
    X_train_enhanced, 
    y_train_resampled,
    eval_metric=['auc', 'recall'],
    eval_set=[(X_test_enhanced, y_test)],
)

lgbc_cost_sensitive
```

```python
# Standard prediction
y_pred_cs = lgbc_cost_sensitive.predict(X_test_enhanced)
y_prob_pred_cs = lgbc_cost_sensitive.predict_proba(X_test_enhanced)
y_prob_pred_cs = [x[1] for x in y_prob_pred_cs]

print("Cost-sensitive model evaluation:")
compute_evaluation_metric(lgbc_cost_sensitive, X_test_enhanced, y_test, y_pred_cs, y_prob_pred_cs)
```

## 4. Threshold Optimization for Better Recall

Let's find the optimal threshold that maximizes recall while maintaining acceptable precision.

```python
from sklearn.metrics import precision_recall_curve, f1_score

# Generate precision-recall curve
precision, recall, thresholds = precision_recall_curve(y_test, y_prob_pred_cs)

# Calculate F1 score for each threshold
f1_scores = []
for i in range(len(precision)):
    if precision[i] > 0 and recall[i] > 0:  # Avoid division by zero
        f1 = 2 * (precision[i] * recall[i]) / (precision[i] + recall[i])
        f1_scores.append((thresholds[i-1] if i > 0 else 0, f1, precision[i], recall[i]))

# Sort by F1 score
f1_scores.sort(key=lambda x: x[1], reverse=True)
optimal_threshold, best_f1, best_precision, best_recall = f1_scores[0]

print(f"Optimal threshold: {optimal_threshold:.4f}")
print(f"Best F1 Score: {best_f1:.4f}")
print(f"Precision at this threshold: {best_precision:.4f}")
print(f"Recall at this threshold: {best_recall:.4f}")

# Plot precision-recall curve
plt.figure(figsize=(10, 6))
plt.plot(recall, precision, 'b-', linewidth=2)
plt.axvline(x=best_recall, color='r', linestyle='--')
plt.axhline(y=best_precision, color='r', linestyle='--')
plt.xlabel('Recall', fontsize=14)
plt.ylabel('Precision', fontsize=14)
plt.title('Precision-Recall Curve with Optimal Threshold', fontsize=16)
plt.grid(True)
plt.show()
```

```python
# Apply custom threshold to predictions
y_pred_optimized = (y_prob_pred_cs >= optimal_threshold).astype(int)

# Evaluate with optimized threshold
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_auc_score

print("\nEvaluation with optimized threshold:")
print("\nAccuracy Score: ", accuracy_score(y_test, y_pred_optimized))
print("\nAUC - ROC Score: ", roc_auc_score(y_test, y_prob_pred_cs))
print("\nClassification Report:\n", classification_report(y_test, y_pred_optimized))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_optimized))
```

## 5. Ensemble Approach - Stacking Multiple Models

Let's create an ensemble of our best models to further improve performance.

```python
from sklearn.ensemble import VotingClassifier, StackingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier

# Train base models on resampled data with enhanced features
print("Training Random Forest...")
rfc_enhanced = RandomForestClassifier(random_state=42, n_estimators=100, class_weight='balanced')
rfc_enhanced.fit(X_train_enhanced, y_train_resampled)

print("Training XGBoost...")
xgb_enhanced = XGBClassifier(random_state=42, scale_pos_weight=10)  # scale_pos_weight balances positive/negative weights
xgb_enhanced.fit(X_train_enhanced, y_train_resampled)
```

```python
from sklearn.ensemble import VotingClassifier, StackingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier

# Train base models on resampled data with enhanced features
print("Training Random Forest...")
rfc_enhanced = RandomForestClassifier(random_state=42, n_estimators=100, class_weight='balanced')
rfc_enhanced.fit(X_train, y_train)

print("Training XGBoost...")
xgb_enhanced = XGBClassifier(random_state=42, scale_pos_weight=10)  # scale_pos_weight balances positive/negative weights
xgb_enhanced.fit(X_train, y_train)
```

```python
# Create a stacking ensemble
estimators = [
    ('rf', rfc_enhanced),
    ('xgb', xgb_enhanced),
    ('lgbm', lgbc_cost_sensitive)
]

stacking_clf = StackingClassifier(
    estimators=estimators,
    final_estimator=LogisticRegression(class_weight='balanced'),
    cv=5,
    stack_method='predict_proba'
)

# Train the stacking model
stacking_clf.fit(X_train_enhanced, y_train_resampled)

# Make predictions with the ensemble
y_pred_stack = stacking_clf.predict(X_test_enhanced)
y_prob_pred_stack = stacking_clf.predict_proba(X_test_enhanced)[:, 1]

# Evaluate the ensemble
print("\nEnsemble Model Evaluation:")
compute_evaluation_metric(stacking_clf, X_test_enhanced, y_test, y_pred_stack, y_prob_pred_stack)
```
